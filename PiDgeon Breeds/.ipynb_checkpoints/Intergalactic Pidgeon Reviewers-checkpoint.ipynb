# Part 1

Load Table and Mingle with df

```python
df = pd.read_csv(r"Matrix.csv")
page_count = int(df.size**.5)
df.set_index('page', inplace=True)
df.head(page_count + 1)
```

```python
df.fillna(0, inplace=True)
df = df.astype(float)
df.head(page_count + 1)
```

```python
df_test = df.copy()
df_test.head(page_count + 1)
```

## a) Create a column-stochastic matrix in the form of a DF


```python
def colum_stoch(df_matrix):
    for i in df_matrix.columns: 
        summ = 0
        for j in df_matrix.columns:
            summ += float(df_matrix.loc[j, i])
        for k in df_matrix.columns:
            if summ == 0:
                df_matrix.loc[k,i] = 1/page_count
            else:
                df_matrix.loc[k,i] /= summ
    return df_matrix
colum_stoch(df_matrix=df_test).head(page_count + 1)
```

## a) Go Go Matrix Rangers

P.S. Take liberty in assigning the p values to test the effects, I'll just leave it at 0.85 for generality.

```python
def mat_rang(df, p):
    matrix = df.transpose().to_numpy()
    A = p * matrix + (1 - p)/page_count
    return A.transpose()
A = mat_rang(df=df_test, p= 0.85)
e_dataframe = pd.DataFrame(A)
e_dataframe.head(page_count + 1)
```

## Get Eigenvalues and Eigenvector
Though they don't match and I had to figure out how to match them manually cuz like why wouldn't they scramble everything up

```python
def near(a, b, rtol = 1e-5, atol = 1e-8):
    return np.abs(a-b)<(atol+rtol*np.abs(b))
def get_eig_1(mat_config):
    eig_val, eig_vec = eig(mat_config)
    eig_vec = eig_vec.T
    for val, vec in zip(eig_val, eig_vec):
        assert np.allclose(np.dot(mat_config, vec), val*vec)
    eig_1 = eig_vec[near(eig_val, 1.0)][0]
    return eig_1
eig_1 = get_eig_1(mat_config=A)
eig_1
```

## scale that bad boy up
Make sure the first norm of the vector is equal to 1.

```python
def scale(v):
    summ = 0
    scaled_v = []
    for i in v:
        summ += abs(i)
    for j in v:
        scaled_v.append(j / summ)
    for k,m in enumerate(scaled_v):
        scaled_v[k] = abs(np.round(m,2))
    return scaled_v
scaled_eig_1 = scale(eig_1)
def const(v, scaled_v):
    return v[0] / scaled_v[0]
coeff = const(eig_1, scaled_eig_1)
scaled_eig_1
```

## Rank piDgeons cuz what is life at this point

```python
def ranking(scaled_v):
    dic = {}
    for i, col in enumerate(df.columns):
        dic[col] = scaled_v[i]
    return dic
dict_ranking = ranking(scaled_eig_1)
dict_ranking
```

# Part 2 : Mid-Max Life Crisis

```python
df_opt = df.copy()

#mid-maxing - add connection form 't' to 'c', which seems to be the highest ranked page.
df_opt.loc['columbidae','telegraph'] = 1.0

#converting, scaling, comparing
def produce_ranking(df_mid_max):
    df_mid_max = colum_stoch(df_mid_max)
    A_opt = mat_rang(df=df_mid_max, p= 0.85)
    eig_opt = get_eig_1(mat_config=A_opt)
    scaled_eig_opt = scale(eig_opt)
    dict_ranking_opt = ranking(scaled_eig_opt)
    return dict_ranking_opt

# visualize data
#check = pd.DataFrame(columns = df.columns, data=A_opt)

#check if mid-maxing was successful
dict_ranking = produce_ranking(df_mid_max=df_opt)
dict_ranking
```

## Write algorithm to find the optimal connected system by changing exactly one entry
Did it woohooo

```python
all = {}
all[('orig','orig')] = 0.11

def iterate_like_hell(df):
    df_change = df.copy()
    cols = df.columns
    for i in cols:
        for j in cols:
            if i is j:
                continue
            if df.loc[i,j] == 0:
                df_change.loc[i,j] = 1
            else:
                df_change.loc[i,j] = 0
            dict_ranking = produce_ranking(df_mid_max=df_change)
            all[(j,i)] = dict_ranking['columbidae']
            df_change.loc[i,j] = df.loc[i,j]
    return dict_ranking

dict_ranking = iterate_like_hell(df=df)
def extract(dic):
    c={}
    for key,value in dic.items():
        if 'columbidae' in key:
            c[key] = value
    return c
c = extract(dic=all)
opt_from,opt_to = max(c,key=c.get)
```
